```python
import pandas as pd
import requests
from io import StringIO
import numpy as np
```

```python
res1 = requests.get('https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/refs/heads/main/day34-handling-date-and-time/messages.csv').text
res2 = requests.get('https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/refs/heads/main/day34-handling-date-and-time/orders.csv').text
dfDate = pd.read_csv(StringIO(res2))
dfTime = pd.read_csv(StringIO(res1))
```

```python
dfDate.head()
```

```python
dfDate.info()
```

```python
dfDate['date'] = pd.to_datetime(dfDate['date'])
```

```python
dfDate.info()
```

```python
dfDate['year'] = dfDate['date'].dt.year
dfDate['month'] = dfDate['date'].dt.month
dfDate['month_name'] = dfDate['date'].dt.month_name()
dfDate['day'] = dfDate['date'].dt.day
dfDate['dayName'] = dfDate['date'].dt.day_name()
dfDate['week'] = dfDate['date'].dt.weekday
dfDate['dayOfWeek'] = dfDate['date'].dt.dayofweek
dfDate['quarter'] = dfDate['date'].dt.quarter
dfDate['semester'] = np.where(dfDate['quarter'].isin([1,2]), 1, 2)
dfDate['isWeekend'] = np.where(dfDate['dayName'].isin(['Sunday', 'Saturday']), True, False)

dfDate.head()
```

```python
import datetime 
today = datetime.datetime.today()
today
```

```python
print('difference = ',today - dfDate['date'])
```

```python
print('days = ',(today - dfDate['date']).dt.days)
```

```python
np.round((today - dfDate['date']) / np.timedelta64(1, 'D') / 30,0)
```
