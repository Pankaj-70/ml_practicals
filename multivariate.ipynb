```python
import seaborn as sns
import pandas as pd
import requests
from io import StringIO
```

```python
iris = sns.load_dataset('iris')
flights = sns.load_dataset('flights')
tips = sns.load_dataset('tips')
response = requests.get('https://raw.githubusercontent.com/datasciencedojo/datasets/refs/heads/master/titanic.csv').text
titanic = pd.read_csv(StringIO(response))
```

```python
print(titanic.columns)
print(iris.columns)
print(tips.columns)
print(flights.columns)
```

```python
sns.scatterplot(data = tips, x = tips['total_bill'], y = tips['tip'], hue = tips['sex'], style = tips['day'], size = tips['size'])
```

```python
sns.barplot(data = titanic, x = titanic['Pclass'], y = titanic['Survived'], hue = titanic['Sex'])
```

```python
sns.boxplot(data = titanic['Age'])
```

```python
sns.displot(data=titanic, x=titanic['Age'], hue=titanic['Survived'], kind='kde')   
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```

```python
sns.heatmap(pd.crosstab(flights['year'], flights['passengers']))
```

```python
(titanic.groupby('Pclass')['Survived'].mean()*100).plot(kind='bar')
```

```python
sns.pairplot(iris, hue = 'species')
```

```python
sns.heatmap(flights.pivot_table(values='passengers', index='month', columns='year'))
```
