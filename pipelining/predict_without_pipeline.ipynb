```python
import pickle
import numpy as np
```

```python
ohe_sex = pickle.load(open('models/ohe_sex.pkl', 'rb'))
ohe_emb = pickle.load(open('models/ohe_emb.pkl', 'rb'))
clf = pickle.load(open('models/clf.pkl', 'rb'))
```

```python
test_input = np.array([2, 'male', 31.0, 0, 0, 10.5, 'S'],dtype=object).reshape(1,7)
test_input
```

```python
test_input_sex = ohe_sex.transform(test_input[:,1].reshape(1,1)).toarray()
test_input_emb = ohe_emb.transform(test_input[:,-1].reshape(1,1)).toarray()
test_input_age = test_input[:,2].reshape(1,1)
```

```python
test_input_transformed = np.concatenate((test_input[:,[0, 3, 4, 5]], test_input_age, test_input_sex, test_input_emb), axis = 1)
```

```python
clf.predict(test_input_transformed)
```
